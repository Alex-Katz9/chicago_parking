## Functions used for cleaning and some feature engineering

```python
def violation_codes(table):
    df = pd.DataFrame()
    #Simplify codes to fewer subsections
    df = table[['license_plate_state','license_plate_number','license_plate_type','violation_code','fine_level1_amount','hearing_disposition','ticket_queue','geocoded_lng','geocoded_lat','violation_code']]
    df['code'] = [code[:7] for code in table['violation_code']]
    df.drop(columns='violation_code', inplace=True)
    print(df.shape)
    # use pd.get_dummies to join the new columns with your original dataframe
    df = pd.get_dummies(df, columns=['code'], prefix='code', prefix_sep='_') 
    print(df.shape)
    return df
```

```python
def clean_data(df):
    table = pd.DataFrame()
    table = df
    table = table.convert_dtypes()
    table = table[table.license_plate_type == 'PAS']
    table.drop(columns='license_plate_type',inplace=True)
    
    #Create a feature for how many times a plate has been ticketed, replace the generic plate frequency with 1
    table['freq'] = table.groupby('license_plate_number')['license_plate_number'].transform('count')
    table['freq'].loc[(table['freq'] > 1000)] = 1
    table.drop(columns='license_plate_number',inplace=True)
    
    #Mark as paid or not
    #df['status'] = table['ticket_queue'].transform(lambda x: np.where(x=='Paid',1,0))
    table.dropna(subset=['ticket_queue'],inplace=True)
    table['status'] = (table.ticket_queue == 'Paid').astype(int)
    table['status'] = (table.status == 0).astype(int)
    table.drop(columns='ticket_queue',inplace=True)
     
    #Place state as IL or other
    table.dropna(subset=['license_plate_state'],inplace=True)
    table['state'] = (table.license_plate_state == 'IL').astype(int)
    table.drop(columns='license_plate_state',inplace=True)

    #Label outcome of hearing as liable or not liable
    table.hearing_disposition.fillna('None',inplace=True)
    table['hearing'] = (table.hearing_disposition == 'Liable').astype(int)
    table.drop(columns='hearing_disposition',inplace=True)
    
    #Include value of ticket
    table['value'] = table['fine_level1_amount']
    table.drop(columns='fine_level1_amount',inplace=True)
    
    #Place geo-location data in table
    table.rename(columns={'geocoded_lng': 'longitude', 'geocoded_lat': 'latitude'},inplace=True)
    table = table[table['longitude'] < -85]
    table = table[table['latitude'] > 40]
    table.dropna(subset=['longitude','latitude'],inplace=True)
    
    return table
```

## Cleaning and splitting data into train and test, to be used in other notebooks
